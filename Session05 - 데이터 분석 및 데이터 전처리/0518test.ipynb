# 1. 라이브러리 선언

```python
import pandas as pd
import numpy as np
```

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv",encoding="ms949")
```

# 2. 데이터 불러오기

```python
selloutData.dtypes
selloutData = selloutData.astype({
    "YEARWEEK":str})

# 불량데이터 처리
selloutData["QTY_NEW"]= np.where(selloutData.QTY<0,0,selloutData.QTY)
```

# 문제 1번

```python
# 연도별 데이터 조회를 위한 year / week 컬럼 형성
yearAndWeek = 4
selloutData["YEAR"]=selloutData.YEARWEEK.str[:yearAndWeek]
selloutData["WEEK"]=selloutData.YEARWEEK.str[yearAndWeek:]
```

```python
# a01지역, 2015년 데이터만 조회
selectedData = selloutData.loc[(selloutData.REGIONID=="A01")&(selloutData.YEAR=="2015")]
```

```python
# 필요한 데이터만 뽑기
groupkey=["REGIONID","PRODUCT","YEAR"]

# 평균 데이터 컬럼 생성
groupData = selectedData.groupby(by=groupkey)["QTY_NEW"].agg(["mean"])

# 컬럼명 변경
groupData.rename(columns={"mean":"QTY_MEAN"},inplace=True)
groupData.reset_index(inplace=True)

# 내림차순 정렬
sortkey=["QTY_MEAN"]
groupData.sort_values(by=sortkey,ascending=[False],inplace=True)

groupData
```
