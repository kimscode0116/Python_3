# 라이브러리 선언

```python
import pandas as pd
import numpy as np
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")

customerData
```

# 정렬하기

```python
customerData.columns
```

```python
sortKey = ["EMI","AVGPRICE"]

sortedData = customerData.sort_values(by=sortKey,
                                     ascending=[True,True],
                                    inplace=False, ignore_index=True)

sortedData
```

# 집계함수

### 예제 1)

```python
groupKey = ["EMI"]

meanValue = sortedData.\
    groupby(by=groupKey)\
    ["AVGPRICE"].\
    agg(["mean"])

meanValue.rename(columns={"mean":"MEAN_AVGPRICE"},
                inplace=True)

# 컬럼명 변경(mean => MEAN_AVGPRICE)
meanValue.reset_index(inplace=True)
meanValue
```

### 예제 2)

```python
meanValue2 = sortedData.\
    groupby(by=groupKey)\
    ["AVGPRICE"].\
    agg(["count"])
```

```python
meanValue2.rename(columns={"count":"SIZE_CS"},
                 inplace=True)
```

```python
meanValue2.reset_index(inplace=True)

meanValue2
```
